```python
# Effect of living/working near the gym on churn
df['Partner'] = df['Partner'].astype(str)
df['Churn'] = df['Churn'].astype(str)

# Check for unique values in 'Partner' and 'Churn' columns
print(df['Partner'].unique())
print(df['Churn'].unique())

sns.countplot(x='Near_Location', hue='Churn', data=df)
plt.title('Churn by Proximity to Gym')
plt.show()
```

```python
# Ensure 'Age' and 'Churn' columns exist and are correctly named
if 'Age' in df.columns and 'Churn' in df.columns:
    # Create age groups
    df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 30, 40, 50, 60, 100], labels=['0-18', '19-30', '31-40', '41-50', '51-60', '60+'])
    
    # Plot churn rate by age group
    plt.figure(figsize=(10, 6))
    sns.countplot(x='AgeGroup', hue='Churn', data=df, palette='pastel')
    plt.title('Churn Rate by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.show()
else:
    print("Columns 'Age' or 'Churn' do not exist in the DataFrame.")
```
